## Performing Analysis of Meteorological Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading a csv file


```python
df = pd.read_csv('D:\DS\Ds seven project\project 1\weatherhistory.csv')
df.head(8)
```

## Now I am going to remove any unnecessary column so that the dataframe looks more relevant

```python
df.drop(['Daily Summary','Loud Cover'],axis = 1, inplace  = True)
df
```

## Now I am going to check how many null entries are there

```python
null_val = df[df.isnull().any(axis=1)]
null_val
```

## I am going to fill the null values with "null"

```python
df.fillna('null', axis = 1,inplace = True)
df
```

## Now we can see that all the null values are replaced

```python
df.isnull().sum()
```

```python
df[['Formatted Date']]
```

## Resampling data into monthly 

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True,dayfirst = True)
df2=df.set_index('Formatted Date')
df2 = df2.resample('M').mean()
df2
```

## Visualizing relations through plotting

```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
sns.relplot(x = 'Temperature (C)', y = 'Humidity',data = df2,kind = 'line', color = 'pink').set(title = "Relation between Humidity and Temperature")
plt.show()
```

## Plotting the relation between Apparent temperature and Humidity

```python
plt.figure(figsize=(15,8))
sns.relplot(x='Apparent Temperature (C)',y='Humidity',color='red', hue = 'Summary', data = df,col = 'Summary', col_wrap = 3)

plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Apparent Temperature and Humidity with year")
plt.plot(df2["Humidity"],label='Humidity',color='red')
plt.plot(df2['Apparent Temperature (C)'],label='Apparent Temperature', color='aqua',marker = "*")
plt.xlabel("Year",fontsize="10")
plt.legend(loc=5,fontsize=10)
plt.show()
```

## Relation between Humidity and Wind Speed (km/h)

```python
sns.relplot(x = 'Wind Speed (km/h)', y = 'Humidity', data = df2, kind = 'line', color = 'purple', ci = None).set(title = "Relation between Humidity and Wind Speed (km/h)")

plt.show()
```
